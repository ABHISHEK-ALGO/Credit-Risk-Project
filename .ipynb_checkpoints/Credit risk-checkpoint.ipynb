# DATA PREPROCESSING

```python
df1.info()
```

```python
# from business we come to know that -9999 appearing in the dataset are null values so we will try to look into that and impute  or remove them accordingly.

df1 = df1.loc[df1['Age_Oldest_TL']  !=-99999]
df1.shape
```

```python
# now lets see df2 
df2.info()
```

```python
# looking for -99999 in this also,we see that it is present in large number so we have decided to remove the column if it has more than 100000 rows belong to -99999 becuse that col will become useless and 
# if -99999 in rows are less than 10000  in a rows then we will remove those rows
df2
```

```python
columns_to_remove = []
for i in df2.columns:
    if df2.loc[df2[i]== -99999].shape[0]>10000:
        columns_to_remove.append(i)
```

```python
columns_to_remove
```

```python
# dropping these columns
df2 = df2.drop(columns_to_remove,axis=1)
```

```python
df2.shape
```

```python
for i in df2.columns:
    df2 = df2.loc[df2[i]!=-99999]
```

```python
df2.shape
```

```python
df1.isna().sum()
df2.isna().sum()
```

```python
# hence null values are removed now lets merge the two datasets besed on common column
for i in list(df1.columns):
    if i in list(df2.columns):
        print(i)
```

```python
# merging on PROSPECTED column
df = pd.merge(df1,df2,how='inner',left_on = ['PROSPECTID'],right_on =['PROSPECTID'])
df.isna().sum()
```

so we have removed null values since models do not treat null values , although there are many categorical data which we need to handle.
